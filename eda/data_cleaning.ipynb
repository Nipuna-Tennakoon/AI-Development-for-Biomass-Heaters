```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../dataset/final_df_001.csv')
df.head()
```

```python
df.tail(25)
```

```python
selected_features = df.columns.tolist()
for i in selected_features:
    print(i)
```

```python
# Convert 'Date Time' column to datetime objects for proper plotting
df['Date Time'] = pd.to_datetime(df['Date Time'])

# List of features to plot
features_to_plot = [
    'Total Load Output', 'Heater Heat Output', 'T.F Heater Outlet Flow Transmitter',
    'T.F. Heater Outlet Temperature - 1', 'T.F. Heater Outlet Temperature - 2 (Modulation)',
    'T.F. Heater Outlet Temperature - 3 (Safety)', 'T.F. Heater Serpentine Coil Inlet Temperature No.1',
    'T.F. Heater Serpentine Coil Inlet Temperature No.2', 'T.F. Heater Serpentine Coil Outlet Temperature',
    'Heater Outlet Flue Gas O2 Analyzer', 'T.F. Heater Outlet Flue Gas Temperature',
    'Furnace Exit Temperature -1', 'Furnace Exit Temperature -2'
]

# Create a 5x3 subplot grid
fig, axs = plt.subplots(nrows=5, ncols=3, figsize=(20, 25))
# Flatten the axes array for easy iteration
axs = axs.flatten()

# Plot each feature on a separate subplot
for i, feature in enumerate(features_to_plot):
    if feature in df.columns:
        axs[i].plot(df['Date Time'], df[feature])
        axs[i].set_title(feature, fontsize=10)
        axs[i].tick_params(axis='x', rotation=45)
    else:
        print(f"Column '{feature}' not found in DataFrame.")

# Hide any unused subplots
for i in range(len(features_to_plot), len(axs)):
    fig.delaxes(axs[i])

# Adjust layout to prevent labels from overlapping
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation matrix for numerical columns
correlation_matrix = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
# --- Feature Engineering: Averaging Related Columns ---

# 1. Average T.F. Heater Outlet Temperatures
heater_outlet_temp_cols = [
    'T.F. Heater Outlet Temperature - 1',
    'T.F. Heater Outlet Temperature - 2 (Modulation)',
    'T.F. Heater Outlet Temperature - 3 (Safety)'
]
df['Average T.F. Heater Outlet Temperature'] = df[heater_outlet_temp_cols].mean(axis=1)

# 2. Average T.F. Heater Serpentine Coil Inlet Temperatures
serpentine_inlet_temp_cols = [
    'T.F. Heater Serpentine Coil Inlet Temperature No.1',
    'T.F. Heater Serpentine Coil Inlet Temperature No.2'
]
df['Average T.F. Heater Serpentine Coil Inlet Temperature'] = df[serpentine_inlet_temp_cols].mean(axis=1)

# 3. Average Furnace Exit Temperatures
furnace_exit_temp_cols = [
    'Furnace Exit Temperature -1',
    'Furnace Exit Temperature -2'
]
df['Average Furnace Exit Temperature'] = df[furnace_exit_temp_cols].mean(axis=1)

# --- Cleanup ---
# Drop the original columns that have been averaged
cols_to_drop = heater_outlet_temp_cols + serpentine_inlet_temp_cols + furnace_exit_temp_cols + ['Total Load Output']
df_cleaned = df.drop(columns=cols_to_drop)

print("Created new averaged columns and dropped original columns.")
print("New DataFrame shape:", df_cleaned.shape)
print("\nFirst 5 rows of the updated DataFrame:")
df_cleaned.head()
```

```python
# --- Visualization of Cleaned Features ---

# Note: We are now using the df_cleaned DataFrame created in the previous step.

# List of new and existing features to plot
features_to_plot = [
    'Heater Heat Output', 'T.F Heater Outlet Flow Transmitter',
    'T.F. Heater Serpentine Coil Outlet Temperature', 'Heater Outlet Flue Gas O2 Analyzer',
    'T.F. Heater Outlet Flue Gas Temperature', 'Average T.F. Heater Outlet Temperature',
    'Average T.F. Heater Serpentine Coil Inlet Temperature', 'Average Furnace Exit Temperature'
]

# Create a 3x3 subplot grid, which is a better fit for 9 plots
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(20, 15))
# Flatten the axes array for easy iteration
axs = axs.flatten()

# Plot each feature on a separate subplot from the cleaned dataframe
for i, feature in enumerate(features_to_plot):
    if feature in df_cleaned.columns:
        axs[i].plot(df_cleaned['Date Time'], df_cleaned[feature])
        axs[i].set_title(feature, fontsize=12)
        axs[i].tick_params(axis='x', rotation=45)
    else:
        # This will now correctly check in df_cleaned
        print(f"Column '{feature}' not found in the cleaned DataFrame.")

# Hide any unused subplots
for i in range(len(features_to_plot), len(axs)):
    fig.delaxes(axs[i])

# Adjust layout to prevent labels from overlapping
plt.tight_layout()
plt.show()
```

```python
# --- Outlier Handling using the IQR Method ---

# Make a copy to preserve the original df_cleaned
df_no_outliers_iqr = df_cleaned.copy()

# List of features to clean
features_to_clean = [
    'Heater Heat Output', 'T.F Heater Outlet Flow Transmitter',
    'T.F. Heater Serpentine Coil Outlet Temperature', 'Heater Outlet Flue Gas O2 Analyzer',
    'T.F. Heater Outlet Flue Gas Temperature', 'Average T.F. Heater Outlet Temperature',
    'Average T.F. Heater Serpentine Coil Inlet Temperature', 'Average Furnace Exit Temperature'
]

print("Processing outliers using the IQR method...")
# Loop through each specified feature to handle outliers
for feature in features_to_clean:
    if feature in df_no_outliers_iqr.columns:
        # Calculate Q1, Q3, and IQR
        Q1 = df_no_outliers_iqr[feature].quantile(0.25)
        Q3 = df_no_outliers_iqr[feature].quantile(0.75)
        IQR = Q3 - Q1
        
        # Define the outlier bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Calculate the median to replace outliers with (more robust than mean)
        median_val = df_no_outliers_iqr[feature].median()
        
        # Identify outliers
        outliers = df_no_outliers_iqr[
            (df_no_outliers_iqr[feature] < lower_bound) | (df_no_outliers_iqr[feature] > upper_bound)
        ]
        
        # Replace outliers with the median value
        df_no_outliers_iqr[feature] = df_no_outliers_iqr[feature].apply(
            lambda x: median_val if x < lower_bound or x > upper_bound else x
        )
        
        print(f"- {feature}: Found and replaced {len(outliers)} outliers.")
    else:
        print(f"- WARNING: Column '{feature}' not found in the DataFrame.")

print("\nOutlier processing complete.")
print("\nFirst 5 rows of the DataFrame with outliers handled using IQR:")
df_no_outliers_iqr.head()
```

```python
# --- Visualization of Cleaned Features ---

# Note: We are now using the df_cleaned DataFrame created in the previous step.

# List of new and existing features to plot
features_to_plot = [
    'Heater Heat Output', 'T.F Heater Outlet Flow Transmitter',
    'T.F. Heater Serpentine Coil Outlet Temperature', 'Heater Outlet Flue Gas O2 Analyzer',
    'T.F. Heater Outlet Flue Gas Temperature', 'Average T.F. Heater Outlet Temperature',
    'Average T.F. Heater Serpentine Coil Inlet Temperature', 'Average Furnace Exit Temperature'
]

# Create a 3x3 subplot grid, which is a better fit for 9 plots
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(20, 15))
# Flatten the axes array for easy iteration
axs = axs.flatten()

# Plot each feature on a separate subplot from the cleaned dataframe
for i, feature in enumerate(features_to_plot):
    if feature in df_no_outliers_iqr.columns:
        axs[i].plot(df_no_outliers_iqr['Date Time'], df_no_outliers_iqr[feature])
        axs[i].set_title(feature, fontsize=12)
        axs[i].tick_params(axis='x', rotation=45)
    else:
        # This will now correctly check in df_cleaned
        print(f"Column '{feature}' not found in the cleaned DataFrame.")

# Hide any unused subplots
for i in range(len(features_to_plot), len(axs)):
    fig.delaxes(axs[i])

# Adjust layout to prevent labels from overlapping
plt.tight_layout()
plt.show()
```
