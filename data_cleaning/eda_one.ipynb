```python
dataset_dir = "../dataset"
sample_xlsx = "August/01.08.2024-02.08.2024.xlsx"
file_path = os.path.join(dataset_dir, sample_xlsx)
df_1 = pd.read_excel(file_path, sheet_name="Group-1", skiprows=7)
df_2 = pd.read_excel(file_path, sheet_name="Group-2", skiprows=7)
df_3 = pd.read_excel(file_path, sheet_name="Group-3", skiprows=7)
df_4 = pd.read_excel(file_path, sheet_name="Group-4", skiprows=7)
```

```python
file_paths = glob.glob("/home/nipuna/Documents/MSC-All/AI-Development-for-Biomass-Heaters/dataset/August/*.xlsx")
aug_list = []
sep_list = []   
for file_path in file_paths:
    if os.path.isfile(file_path) and file_path.endswith('08.2024.xlsx'):
        aug_list.append(file_path)
    elif os.path.isfile(file_path) and file_path.endswith('09.2024.xlsx'):
        sep_list.append(file_path)
```

```python
aug_date_list = []
for file_path in aug_list:
    date = file_path.split('/')[-1].rsplit('.',1)[0]
    aug_date_list.append(date)  
sorted_aug_date_list = sorted(aug_date_list)
print(sorted_aug_date_list)
```

```python
df_aug_1 = pd.DataFrame()
df_aug_2 = pd.DataFrame()
df_aug_3 = pd.DataFrame()
df_aug_4 = pd.DataFrame()

for date_to_find in sorted_aug_date_list:
    matching_files = [file for file in aug_list if date_to_find in file]
    df_aug_1 = pd.concat([df_aug_1, pd.read_excel(matching_files[0], sheet_name="Group-1", skiprows=7).drop(0,axis=0)], ignore_index=True)
    df_aug_2 = pd.concat([df_aug_2, pd.read_excel(matching_files[0], sheet_name="Group-2", skiprows=7).drop(0,axis=0)], ignore_index=True)
    df_aug_3 = pd.concat([df_aug_3, pd.read_excel(matching_files[0], sheet_name="Group-3", skiprows=7).drop(0,axis=0)], ignore_index=True)
    df_aug_4 = pd.concat([df_aug_4, pd.read_excel(matching_files[0], sheet_name="Group-4", skiprows=7).drop(0,axis=0)], ignore_index=True)
    
df_aug_1["Date Time"] = pd.to_datetime(df_aug_1["Date Time"])
df_aug_2["Date Time"] = pd.to_datetime(df_aug_2["Date Time"])
df_aug_3["Date Time"] = pd.to_datetime(df_aug_3["Date Time"])
df_aug_4["Date Time"] = pd.to_datetime(df_aug_4["Date Time"])
```

```python
sep_date_list = []
for file_path in sep_list:
    date = file_path.split('/')[-1].rsplit('.',1)[0]
    sep_date_list.append(date)  
sorted_sep_date_list = sorted(sep_date_list)
print(sorted_sep_date_list)
```

```python
df_sep_1 = pd.DataFrame()
df_sep_2 = pd.DataFrame()
df_sep_3 = pd.DataFrame()
df_sep_4 = pd.DataFrame()

for date_to_find in sorted_sep_date_list:
    matching_files = [file for file in sep_list if date_to_find in file]
    df_sep_1 = pd.concat([df_sep_1, pd.read_excel(matching_files[0], sheet_name="Group-1", skiprows=7).drop(0,axis=0)], ignore_index=True)
    df_sep_2 = pd.concat([df_sep_2, pd.read_excel(matching_files[0], sheet_name="Group-2", skiprows=7).drop(0,axis=0)], ignore_index=True)
    df_sep_3 = pd.concat([df_sep_3, pd.read_excel(matching_files[0], sheet_name="Group-3", skiprows=7).drop(0,axis=0)], ignore_index=True)
    df_sep_4 = pd.concat([df_sep_4, pd.read_excel(matching_files[0], sheet_name="Group-4", skiprows=7).drop(0,axis=0)], ignore_index=True)
    
df_sep_1["Date Time"] = pd.to_datetime(df_sep_1["Date Time"])
df_sep_2["Date Time"] = pd.to_datetime(df_sep_2["Date Time"])
df_sep_3["Date Time"] = pd.to_datetime(df_sep_3["Date Time"])
df_sep_4["Date Time"] = pd.to_datetime(df_sep_4["Date Time"])
```

```python
df_sep_1 = df_sep_1.sort_values(by=['Date Time'], ascending=[True], ignore_index=True)
df_sep_2 = df_sep_2.sort_values(by=['Date Time'], ascending=[True], ignore_index=True)
df_sep_3 = df_sep_3.sort_values(by=['Date Time'], ascending=[True], ignore_index=True)
df_sep_4 = df_sep_4.sort_values(by=['Date Time'], ascending=[True], ignore_index=True)
```
