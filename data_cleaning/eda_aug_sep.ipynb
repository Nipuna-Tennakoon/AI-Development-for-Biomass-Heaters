```python
df[df.columns[0]] = pd.to_datetime(df[df.columns[0]])
df = df.set_index(df.columns[0]) 

def replace_outliers_iqr_interpolate(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outlier_mask = (series < lower_bound) | (series > upper_bound)
    
    cleaned = series.copy()
    cleaned[outlier_mask] = np.nan
    
    cleaned = cleaned.interpolate(method='time')
    
    cleaned = cleaned.ffill().bfill()
    
    return cleaned

for col in df.columns:
    print(f"Processing column: {col}")
    df[col] = replace_outliers_iqr_interpolate(df[col])

print("Outlier detection and interpolation completed.")
```
